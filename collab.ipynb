```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances

ratings = pd.read_csv("ratings_preprocessed.csv")
# print(ratings)

mat = ratings.pivot(index='userId', columns='id', values='rating')
mat = mat.fillna(0)
mat = mat.values

item_item = 1-pairwise_distances(mat, metric="cosine") #similarity
print(item_item)
mat = ratings.pivot(index='userId', columns='id', values='rating')
mat = mat.fillna(0)
print(mat)
```

```python
item_dict = {}
count=0
for i in mat:
    item_dict[i] = count
    count +=1
```

```python
def make_predictions(user_id, mat):
    for i in mat:
        if mat[i][user_id] == 0:
            pred = 0
            den = 0
            for j in mat:
                try:
                    pred += item_item[item_dict[i]][item_dict[i]]*mat[j][user_id]
                    den += item_item[item_dict[i]][item_dict[j]]
                except:
                    pass
            pred = pred/den
            mat[user_id][i] = pred
    return mat[user_id]

mat1 = make_predictions(1,mat)
```
