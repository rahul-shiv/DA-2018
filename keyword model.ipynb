```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt
import pandas as pd
from ast import literal_eval
import plotly.offline as off
import nltk
from nltk.corpus import wordnet
import numpy as np
import json
meta = pd.read_csv('movies_metadata_preprocessed.csv')
```

```python
with open("keyword_dict.json",'r') as f:
    keyword_dict=json.load(f)
with open("new_keyword_occurences.json",'r') as f:
    new_keyword_occurences=json.load(f)



dict_key=dict()
for a,b in new_keyword_occurences:
    dict_key[a]=b
revenue_dict = dict()

for index,row in meta.iterrows():
    if type(row.keywords)==type(0.0):
#         print(index)
        continue
    for name in literal_eval(row.keywords):
        try:
            revenue_dict[name] = revenue_dict.get(name,0)+(row.revenue)/dict_key[name]
        except:
#             print(name)
            pass
# revenue_dict
```

```python
pred = []
for index,row in meta.iterrows():
#     try:
#         print(row.keywords)
        temp=0
        if type(row.keywords)==type(0.0):
            pred.append(np.nan)
            continue
        for name in literal_eval(row.keywords):
            try:
                temp += revenue_dict[name]/len(literal_eval(row.keywords))
                
            except:
#                 print(name)
                pass
        pred.append(temp)
#         print(temp)
#         break
#     except:
#         pred.append(np.nan)
#         print(row.keywords)
#         pass
```

```python
predar=np.array(pred)
meta['pred']=predar
predictions=meta.iloc[:,[11,19]]
predictions['log revenue']=np.log10(predictions.revenue)
predictions['log predictions']=np.log10(predictions.pred)
predictions.to_csv('predictions.csv')
new_predictions = predictions[~predictions.pred.isnull()]
print((new_predictions['log predictions']-new_predictions['log revenue']).mean()**.5)
```
